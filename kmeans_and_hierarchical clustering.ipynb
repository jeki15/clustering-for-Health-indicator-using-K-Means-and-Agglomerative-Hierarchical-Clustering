### K-Means

```python
from sklearn.cluster import KMeans
```

#### Inertia Method

```python
# finding optimal number of clusters
inertia = []
max_clusters = 11
for n_cluster in range(1, max_clusters):
    kmeans = KMeans(n_clusters=n_cluster, random_state=123, n_init=n_cluster)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)
```

```python
#Elbow Curve
plt.figure(figsize=(8,5))
plt.plot(np.arange(1, max_clusters), inertia, marker='o')
plt.xlabel('Number of cluster')
plt.ylabel('Inertia')
plt.xticks(np.arange(1, max_clusters))
plt.show()
```

Dalam kurva elbow, titik 'belokan' yang terbentuk pada grafik umumnya dianggap sebagai petunjuk jumlah klaster yang sesuai. Meskipun demikian, nilai 'optimal' dari 'k' yang diperoleh dari metode kurva elbow sering kali agak "ambigu" atau tidak sepenuhnya pasti dalam menentukan jumlah klaster ('k') yang terbaik. Pada grafik di atas, pilihan 'k' diatur menjadi 2 atau 3 karena grafik yang dihasilkan menunjukkan adanya belokan dan kemudian level off. Untuk memastikan bahwa 'k' yang dipilih memang optimal, analisis lebih lanjut akan dilakukan menggunakan analisis silhouette.


#### Silhouette Analysis Method

```python
from sklearn.metrics import silhouette_score

# Silhouette Analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:

    # Initialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=500, random_state=123)
    kmeans.fit(scaled_df)
    cluster_labels = kmeans.labels_

    # Silhouette Score
    silhouette_avg = silhouette_score(scaled_df, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))

```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.cluster import KMeans
import numpy as np

# Your scaled data (assuming it's stored in scaled_df)

range_n_clusters = [2, 3, 4, 5]

for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=500, random_state=123)
    kmeans.fit(scaled_df)
    cluster_labels = kmeans.labels_

    # Silhouette Score
    silhouette_avg = silhouette_score(scaled_df, cluster_labels)

    # Davies-Bouldin Index
    db_score = davies_bouldin_score(scaled_df, cluster_labels)

    # Calinski-Harabasz Index
    ch_score = calinski_harabasz_score(scaled_df, cluster_labels)

    # Dunn Index (Note: Scikit-learn doesn't have Dunn Index implementation, you may need to use external libraries or create a custom function)

    print("For n_clusters={0}, Silhouette Score: {1}, Davies-Bouldin Index: {2}, Calinski-Harabasz Index: {3}".format(num_clusters, silhouette_avg, db_score, ch_score))
```


Silhouette score dapat membantu menentukan jumlah cluster yang optimal dalam analisis klaster (clustering). Semakin tinggi silhouette score, semakin baik. Dalam kasus ini, score tertinggi adalah untuk n_clusters=2 dengan nilai silhouette sebesar 0.7884

### Pemilihan Model

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import davies_bouldin_score, silhouette_score, calinski_harabasz_score
from scipy.spatial.distance import pdist, squareform
import numpy as np
import pandas as pd

# List jumlah klaster yang akan diuji
n_clusters = [2, 3, 4]

# List metode klaster
linkage_methods = ['single', 'average', 'complete', 'ward']

# Inisialisasi dictionary untuk menyimpan hasil indeks validitas
results = {'Metode': [], 'Jumlah Klaster': [], 'DB': [], 'CH': [], 'Silhouette':[], 'Dunn': []}

for method in linkage_methods:
    for n in n_clusters:
        # Melakukan klasterisasi menggunakan metode klaster hierarki
        clustering = AgglomerativeClustering(n_clusters=n, linkage=method)
        cluster_labels = clustering.fit_predict(df_hc)

        # Menghitung nilai indeks validitas
        db_score = davies_bouldin_score(df_hc, cluster_labels)
        ch_score = calinski_harabasz_score(df_hc, cluster_labels)  #
        silhouette_score_avg = silhouette_score(df_hc, cluster_labels)

        # Menghitung Dunn Index
        pairwise_distances = squareform(pdist(df_hc))
        min_inter_cluster_distance = np.inf
        max_intra_cluster_diameter = 0

        for i in range(n):
            for j in range(i + 1, n):
                if cluster_labels[i] != cluster_labels[j]:
                    d = np.min(pairwise_distances[np.where(cluster_labels == i)][:, np.where(cluster_labels == j)])
                    if d < min_inter_cluster_distance:
                        min_inter_cluster_distance = d

        for i in range(n):
            max_intra_cluster_distance = np.max(pairwise_distances[np.where(cluster_labels == i)][:, np.where(cluster_labels == i)])
            if max_intra_cluster_distance > max_intra_cluster_diameter:
                max_intra_cluster_diameter = max_intra_cluster_distance

        dunn_index = min_inter_cluster_distance / max_intra_cluster_diameter

        # Menyimpan hasil ke dalam dictionary
        results['Metode'].append(method)
        results['Jumlah Klaster'].append(n)
        results['DB'].append(round(db_score, 4))
        results['CH'].append(round(ch_score, 4))
        results['Silhouette'].append(round(silhouette_score_avg, 4))
        results['Dunn'].append(round(dunn_index, 4))

# Membuat DataFrame dari hasil
results_df_hc = pd.DataFrame(results)

# Menampilkan hasil
print(results_df_hc.to_string(index=False))
```

Dalam menentukan model terbaik, Anda dapat mempertimbangkan beberapa faktor dan metrik validitas clustering yang telah dihitung. Mari kita tinjau beberapa metrik tersebut:

Davies-Bouldin Index (DB): Nilai DB yang lebih rendah menunjukkan clustering yang lebih baik. Dalam kasus ini, metode "ward" dengan jumlah klaster 2 memiliki nilai DB yang relatif rendah (1.0172), menunjukkan bahwa ini bisa menjadi pilihan yang baik.

Calinski-Harabasz Index (CH): Nilai CH yang lebih tinggi menunjukkan clustering yang lebih baik. Metode "ward" dengan jumlah klaster 2 memiliki nilai CH yang tinggi (7.9096), menunjukkan bahwa ini bisa menjadi pilihan yang baik.

Silhouette Score: Nilai Silhouette yang lebih tinggi menunjukkan bahwa objek di satu klaster lebih mirip satu sama lain dan berbeda dari klaster lainnya. Metode "ward" dengan jumlah klaster 2 memiliki nilai Silhouette yang tinggi (0.3388), menunjukkan bahwa ini bisa menjadi pilihan yang baik.

Dunn Index: Dunn Index yang lebih tinggi menunjukkan bahwa klaster lebih terpisah satu sama lain. Metode "ward" dengan jumlah klaster 2 memiliki Dunn Index yang tinggi (inf), yang juga bisa dianggap sebagai indikasi hasil yang baik.

**Berdasarkan nilai metrik evaluasi, tidak ada satu metode pun yang secara konsisten unggul di semua metrik. Namun, jika kita harus memilih satu, terlihat bahwa klaster 3 dari metode Ward Linkage memiliki nilai DB yang rendah, CH yang tinggi, serta Dunn Index yang relatif baik, meskipun Silhouette Score-nya tidak optimal. Tetapi, pemilihan model terbaik sebaiknya juga mempertimbangkan konteks dan tujuan analisis yang ingin dicapai.**

```python
import statsmodels.api as sm
from statsmodels.multivariate.manova import MANOVA

# Variabel dependen
dependent_vars = ['AKB', 'AKI', 'Persentase_Balita_Gizi_Kurang',
       'Persentase_Pemberian_ASI_Eksklusif',
       'Persentase_Kematian_Akibat_Demam_Berdarah_Dengue_DBD',
       'Jumlah_Dokter_Umum_per_100000_penduduk',
       'Jumlah_Bidan_per_100000_penduduk',
       'Jumlah_Perawat_per_100000_penduduk',
       'Jumlah_Puskesmas_per_100000_penduduk',
       'Jumlah_Posyandu_Aktif_per_100000_penduduk',
       'Persentase_Sumber_Air_Minum_Layak',
       'Persentase_Rumah_Tangga_yang_Memiliki_Akses_Terhadap_Sanitasi_Layak',
       'Persentase_Desa_Kelurahan_Universal_Child_Immunization_UCI',
       'Angka_Keberhasilan_Pengobatan_TBC']

# Variabel independent atau faktor
independent_var = 'Cluster_Ward'  # Ganti dengan nama variabel independent yang sesuai

# Pastikan bahwa 'Cluster Ward' adalah tipe kategori
df_hc['Cluster_Ward'] = df_hc['Cluster_Ward'].astype('category')

# Membuat model MANOVA
formula = f'{" + ".join(dependent_vars)} ~ {independent_var}'
manova_model = MANOVA.from_formula(formula, data=df_hc)

# Menjalankan uji MANOVA
manova_results = manova_model.mv_test()

# Menampilkan hasil
print(manova_results)
```

## ANOVA

```python
import statsmodels.api as sm
import pandas as pd
from statsmodels.formula.api import ols

# Variabel independen
dependent_var = 'Cluster_Ward'

# Membuat dataframe kosong untuk hasil ANOVA
anova_results_df = pd.DataFrame(columns=['Variable', 'Sum_Sq', 'df', 'F', 'PR(>F)'])

# Iterasi untuk setiap variabel dependen
for var in dependent_vars:
    # Membuat formula untuk OLS
    formula = f"{var} ~ C({dependent_var})"

    # Melakukan OLS
    model = ols(formula, data=df_hc).fit()

    # Mengambil hasil ANOVA
    anova_result = sm.stats.anova_lm(model, typ=2)

    # Menambahkan hasil ANOVA ke dataframe
    anova_results_df = anova_results_df.append({
        'Variable': var,
        'Sum_Sq': anova_result['sum_sq'][f'C({dependent_var})'],
        'df': anova_result['df'][f'C({dependent_var})'],
        'F': anova_result['F'][f'C({dependent_var})'],
        'PR(>F)': anova_result['PR(>F)'][f'C({dependent_var})']
    }, ignore_index=True)

# Menampilkan dataframe hasil ANOVA
anova_results_df
```
